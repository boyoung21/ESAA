인공지능 > 머신러닝 > 딥러닝 순으로 포괄적임

## 머신러닝
주어진 데이터를 인간이 먼저 전처리해야 함. 머신러닝 자체는 데이터의 특징을 스스로 추출하지 못하기 때문에, 컴퓨터가 인식할 수 있도록 인간이 준비해 두어야 함
이를 특성 추출이라 하는데, 데이터별로 어떤 특징을 가지고 있는지 찾아내고 그것을 토대로 데이터를 벡터로 변환하는 작업을 의미함. 이때 훈련 데이터 자체가 나쁠 경우 실제 현상의 특성을 제대로 반영할 수 없으므로 실제 데이터의 특징이 잘 반영되고 편향되지 않는 훈련 데이터를 확보하는 것이 중요함.

- 특성추출 -> 인간
- 분류 -> 컴퓨터

또한 머신러닝은 동일한 유형의 데이터 분석을 위한 재사용은 불가능함. 일반적으로 점수 또는 분류 등의 숫자값을 결과로 출력.

## 머신러닝의 모델

머신러닝 학습 단계에서 얻은 최종 결과물로 가설이라고도 함. ex) 입력 데이터의 패턴은 A와 같다

1. 모델(가설) 선택
2. 모델 학습 및 평가
3. 평가를 바탕으로 모델 업데이트

이 세 단계를 반복하면서 모델을 업데이트하고, 이를 최종적으로 문제에 적용하는 과정을 거침.

## 지도학습, 비지도학습, 강화 학습

지도학습: 정답이 무엇인지, 즉 레이블 데이터가 존재하고 이를 활용하여 학습시키는 방법. 명확하게 범주에 따라 분류.

비지도학습: 정답을 알려 주지 않고 특징이 비슷한 데이터를 클러스터링하여 예측하는 학습 방법. 유사도 기반으로 분류.

강화학습: 분류할 수 있는 데이터가 없고, 있다 해도 정답이 없음. 자신의 행동에 대한 보상을 받으며 학습을 진행하는 방식

- 게임이 대표적인 사례. 에이전트가 변화하는 환경에 따라 다른 행동을 취하며, 그 행동에 따라 보상을 얻음. 강화학습은 이런 보상이 커지는 행동을 자주 하도록 하고, 줄어드는 행동은 덜하도록 하여 학습을 진행함.

## 딥러닝

인간이 따로 특성을 추출할 필요 없이, 대량의 데이터를 신경망에 적용하면 컴퓨터가 스스로 분석함. 인간 뇌를 기초로 하여 뉴런과 시냅스 개념을 적용한 것이 특징. 각각의 뉴런은 복잡하게 연결된 수많은 뉴런을 병렬 연산하여 기존 컴퓨터가 수행하지 못했던 음성, 영상 인식 등 처리를 가능하게 함.

- 특성 추출과 분류 모두 컴퓨터가 수행

구현된 알고리즘은 동일한 유형의 데이터를 분석하는 데 재사용되며, 출력은 점수, 텍스트, 소리 등 어떤 것이든 가능함.

딥러닝 모델의 학습 과정은 다음과 같음

1. 데이터 준비
2. 모델(모형) 정의 -> 신경망 생성 단계. 은닉층 개수와 과적합은 서로 상충관계이므로 이를 잘 고려해야 함
3. 모델(모형) 컴파일 -> 이때 옵티마이저와 손실 함수, 활성화 함수 선택 / 훈련 데이터셋의 형태가 연속형이면 MSE, 이진 분류라면 크로스 엔트로피 선택 또한 과적합을 피할 수 있는 활성화 함수 및 옵티마이저 선택이 중요
4. 모델(모형) 훈련 -> 한번에 처리할 데이터 양을 지정. 배치와 에포크 선택이 중요하며, 값의 변화를 시각적으로 표현하여 파라미터와 하이퍼파라미터에 대한 최적값을 찾도록 해야 함. ex) 훈련 데이터셋 1000개에 대한 배치 크기가 20이라면 샘플 단위 20개마다 모델 가중치를 한 번씩 업데이트시킨다는 의미. 이 경우 50번의 가중치가 업데이트됨. 이때 에포크가 10이라면, 결과적으로 가중치가 총 500번 업데이트됨.
5. 모델(모형) 예측

머신러닝과 구분되는 딥러닝의 또 다른 특징은 심층 신경망을 사용한다는 점. 이는 데이터세트의 어떤 특성들이 중요한지 스스로에게 가르쳐 줄 수 있는 기능을 탑재함. 또한, 가중치 값을 업데이트하기 위한 역전파가 중요함. 특히 역전파 계산 과정에서 사용되는 미분이 성능에 영향을 미치는 주요한 요소로, 이때 미분은 오차를 각 가중치로 미분한 것임.

## 지도학습, 비지도학습, 전이학습

지도학습
1. 합성곱 신경망 CNN: 목적에 따라 이미지 분류(이미지 데이터를 유사한 것끼리 분류), 이미지 인식(사진을 분석하여 그 안에 있는 사물의 종류를 인식하는 것), 이미지 분할(영상에서 사물이나 배경 등 객체 간 영역을 픽셀 단위로 구분하는 기술)로 분류 가능.
2. 순환 신경망 RNN: 시간에 따른 데이터가 있을 때 주로 사용하며, 역전파 과정에서 기울기 소멸 문제가 발생하는 단점 존재.
3. LSTM: 순환 신경망의 기울기 소멸 문제를 개선하고자 망각 게이트, 입력 게이트, 출력 게이트 총 3가지를 추가한 알고리즘. 이를 통해 기울기 소멸 문제를 해결

비지도학습
1. 워드 임베딩: 자연어를 컴퓨터가 이해하고 효율적으로 처리하게 만들 수 있도록 단어를 벡터로 표현하는 방식. 단어 의미를 벡터화하는 워드투벡터와 글로브를 가장 많이 사용. 워드 임베딩으로 워드 클라우드 구현 가능.
2. 군집
아무런 정보가 없는 상태에서 데이터를 분류하는 방법으로, 한 클러스터 안의 데이터는 매우 비슷하게 구성하고 다른 클러스터의 데이터와 구분되도록 나누는 것이 목표. 머신러닝의 군집과 유사함. 또, 머신러닝에서 군집화를 처리할 때 딥러닝과 함께 사용하면 모델 성능을 높일 수 있음.

전이학습

사전에 학습이 완료된 모델을 가지고 우리가 원하는 학습에 미세 조정 기법을 이용하여 학습시키는 방법.
- 사전 학습 모델: 풀고자 하는 문제와 비슷하면서 많은 데이터로 이미 학습이 되어 있는 모델. 특성 추출과 미세 조정 기법에서 사전 학습 모델을 활용할 수 있음.

# 모델 정의

파이토치에서 모델을 정의하기 위해서는 모듈을 상속한 클래스를 사용함.

- 계층: 모듈 또는 모듈을 구성하는 한 개의 계층으로 합성곱층, 선형 계층 등이 존재
- 모듈: 한 개 이상의 계층이 모여서 구성된 것으로, 모듈이 모여 새로운 모듈을 만들 수도 있음
- 모델: 최종적으로 원하는 네트워크로, 한 개의 모듈이 모델이 될 수도 있음.

단순 신경망 정의는 다음과 같음. nn.Module을 상속받지 않는 매우 단순한 모델을 만들 때

model = nn.Linear(in_features = 1, out_features = 1, bias = True)

nn.Module()을 상속하여 정의하는 방법

파이토치에서 nn.Module을 상속받는 모델은 기본적으로 __init__() -> 모델에서 사용될 모듈(nn.Linear, nn.Conv2d), 활성화 함수 등을 정의 / forward() 함수를 포함함. -> 모델에서 실행되어야 하는 연산을 정의

# Sequential 신경망을 정의하는 방법

nn.Sequential을 사용하면 __init__()에서 사용할 네트워크 모델들을 정의해 줄 뿐만 아니라, forward() 함수에서는 모델에서 실행되어야 할 계산을 좀 더 가독성 있게 작성 가능

nn.Sequential은 모델의 계층이 복잡할수록 효과가 뛰어남

IMG_0233.jpeg

```python
import torch.nn as nn
class MLP(nn.Module):
  def __init__(self):
    super(MLP, self).__init__()
    self.layer1 = nn.Sequential(
        nn.Conv2d(in_channels=3, out_channels=64, kernel_size=5),
        nn.ReLU(inplace=True),
        nn.MaxPool2d(2))

    self.layer2 = nn.Sequential(
        nn.Conv2d(in_channels=64, out_channels=30, kernel_size=5),
        nn.ReLU(inplace=True),
        nn.MaxPool2d(2))

    self.layer3 = nn.Sequential(
        nn.Linear(in_features=30*5*5, out_features=10, bias=True),
        nn.ReLU(inplace=True))

    def forward(self, x):
      x = self.layer1(x)
      x = self.layer2(x)
      x = x.view(x.shape[0], -1)
      x = self.layer3(x)
      return x
model = MLP()

print("Printing children\n-----------")
print(list(model.children()))
print("\n\nPrinting Modules\n------------")
print(list(model.modules()))
```

# 함수로 신경망을 정의하는 방법

함수로 선언할 경우 변수에 저장해 놓은 계층들을 재사용할 수 있는 장점이 있음. 그러나 모델이 복잡해지는 단점도 존재. 복잡한 모델은 함수를 이용하는 것보다는 nn.Module()을 상속받아 사용하는 것이 편리함

```python
def MLP(in_features = 1, hidden_features = 20, out_features = 1):
  hidden = nn.Linear(in_features = in_features, out_features = hidden_features,
                     bias = True)
  activation = nn.ReLU()
  output = nn.Linear(in_features = hidden_features, out_features = out_features,
                     bias = True)
  net = nn.Sequential(hidden, activation, output)
  return net
```

# 모델 파라미터 정의

모델을 학습하기 전에 필요한 파라미터들을 정의.

1. 손실함수: 학습하는 동안 출력과 실제 값 사이의 오차를 측정하여 모델의 정확성을 측정

- BCELoss: 이진 분류를 위해 사용
- CrossEntropyLoss: 다중 클래스 분류를 위해 사용
- MSELoss: 회귀 모델에서 사용

2. 옵티마이저: 데이터와 손실 함수를 바탕으로 모델의 업데이트 방법을 결정
- optimizer는 step() 메서드를 통해 전달받은 파라미터를 업데이트함
- 모델의 파라미터별로 다른 기준 ex 학습률 을 적용시킬 수 있음
- torch.optim.Optimizer(params, defaults)는 모든 옵티마이저의 기본이 되는 클래스
- zero_grad() 메서드는 옵티마이저에 사용된 파라미터들의 기울기를 0으로 만듦.
- torch.optim.lr_scheduler는 에포크에 따라 학습률 조절 가능

3. 학습률 스케줄러: 미리 지정한 횟수의 에포크를 지날 때마다 학습률을 감소시켜 줌. 학습률 스케줄러를 이용하면 학습 초기에는 빠른 학습을 진행하다가 전역 최소점 근처에 다다르면 학습률을 줄여서 최적점을 찾아갈 수 있도록 해 줌.

- optim.lr_scheduler.LambdaLR: 람다 함수를 이용하여 그 함수의 결과를 학습률로 설정.
- optim.lr_scheduler.StepLR: 특정 단계 (step) 마다 학습률을 감마(gamma) 비율만큼 감소.
- optim .lr_scheduler.MultiStepLR: StepLR과 비슷하지만 특정 단계가 아닌 지정된 에포크에만 감마 비율로 감소.
- optim.lr_scheduler.ExponentialLR: 에포크마다 이전 학습률에 감마만큼 곱.
- optim.lr_scheduler.CosineAnnealingLR: 학습률을 코사인 함수의 형태처럼 변화. 학습률이 커지기도 작아지기도 함.
- optim.lr_scheduler.ReduceLROnPlateau: 학습이 잘되고 있는지 아닌지에 따라 동적으로 학습률을 변화시킬 수 있음.

4. 지표: 훈련과 테스트 단계를 모니터링함
