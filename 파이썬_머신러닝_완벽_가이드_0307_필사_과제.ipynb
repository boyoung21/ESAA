- DataFrame: 판다스의 핵심 객체로, 여러 개의 행과 열로 이뤄진 2차원 데이터를 담는 데이터 구조체

- Index: 개별 데이터를 고유하게 식별하는 Key값 > Series와 DataFrame 모두 Index를 Key값으로 가짐, 고유성이 보장된다면 의미 있는 데이터값 할당도 가능하며 숫자형뿐만 아니라 문자형도 가능, 지정하여 변경 가능

read_csv()
- 칼럼을 ','로 구분한 파일 포맷만 변환할 뿐 아니라, sep 인자를 통해 어떤 필드 구분 문자 기반의 파일 포맷도 DataFrame으로 변환 가능. 기본은 콤마.
- ex) read_csv('파일명', sep='\t') > 탭으로 구분하여 변환한다는 의미
- 별다른 파라미터 지정이 없을 경우, 파일의 맨 처음 로우를 칼럼명으로 인지하고 칼럼으로 변환함.
- 맨 왼쪽에 나타난 파일에 기재되어 있지 않은 로우의 순차값들은 판다스의 Index 객체 값을 의미.

- Null 값을 무시하고 결과값을 내놓기 때문에 Null 값을 포함하고자 한다면 value_counts(dropna=False)로 입력.

- DataFrame > 딕셔너리로의 변환: DataFrame 객체의 to_dict() 메서드를 호출하여 인자로 'list' 입력

DataFrame 데이터 삭제 - drop() 메서드
- drop() 메서드 원형: DataFrame.drop(labels=None, axis = 0, index=None, columns=None, level=None, inplace=False, errors='raise')
- axis에서 행 또는 열을 지정
- labels로 특정 열의 칼럼명을 입력하거나 행의 인덱스 숫자를 지정하여 드롭 수행
- 주로 drop은 칼럼을 가공할 때 사용하며, 행을 삭제하는 경우는 이상치 데이터를 삭제하는 경우에 주로 사용됨
- inplace = True로 설정하면 본래 DataFrame의 데이터를 삭제하며, 기본값은 False이기 때문에 반환된 DataFrame 객체 변수에서만 데이터가 삭제된 채 나타난다. > inplace = True로 설정하면 반환 값이 None이 되기 때문에 반환 값을 다시 자신의 DataFrame으로 할당하면 DataFrame 객체 변수를 아예 None으로 만들어 버림, 다시 할당하지 말고 그냥 drop 메서드 사용한 채로 쓰기
- 여러 개의 칼럼을 삭제하고 싶을 때는 리스트 형태로 삭제하고자 하는 칼럼명을 입력해 labels 파라미터로 입력 ex .drop(['Age_0', 'Age'])

Index 객체
- DataFrame, Series에서 Index 객체만 추출하려면 .index 속성 사용, 반환된 Index 객체의 실제 값은 넘파이 1차원 ndarray로 볼 수 있음
- 한 번 만들어진 데이터프레임 및 시리즈의 인덱스 객체는 변경 불가
- 시리즈 객체에 인덱스 객체가 포함되나, 시리즈 객체에 연산 함수를 적용할 때 인덱스는 연산에서 제외되고 오직 식별용으로만 사용됨
- 데이터프레임 및 시리즈에 reset_index() 메서드를 수행하면 새롭게 인덱스를 연속 숫자 형으로 할당하며 기존 인덱스는 'index'라는 새로운 칼럼명으로 추가됨 > 인덱스가 연속된 정수형자데이터가 아닐 경우에 이를 연속 정수형 데이터로 만들 때 주로 사용됨 ex) value_counts에서 고유값이 식별자 인덱스로 나올 경우, 더 편하게 보기 위해서 연속 숫자형 인덱스를 다시 할당
- 시리즈에 reset_index()를 적용하면 새롭게 숫자형 인덱레가 만들어지고 데이터프레임으로 전환됨
-reset_index(drop = True) > 기존 인덱스는 새로운 칼럼으로 추가되지 않고 삭제됨, 새로운 칼럼이 추가되지 않기 때문에 시리즈 형태를 유지

판다스의 [] 연산자 vs 넘파이의 []의 연산자
- 넘파이의 [] 연산자: 행과 열의 위치, 슬라이싱 범위 등을 지정해 데이터를 가져올 수 있게 함
- 판다스의 [] 연산자: 칼럼명 문자 혹은 칼럼명의 리스트 객체, 인덱스로 변환 가능한 표현식 등 '칼럼 지정 연산자' 역할, 그래서 넘파이의 인덱스 지정처럼 숫자만 입력하면 오류 뜸 /  titanic_df[0:2] or titanic[titanic['Pclass']==3].head(3)와 같은 슬라이싱, 불린 인덱싱 가능

판다스의 iloc[], loc[] - 행과 열을 함께 추출할 때 사용
- iloc[행 위치 정숫값, 열 위치 정숫값]: 위치 기반 인덱싱 > 행과 열의 좌표 위치에 해당하는 값으로 정숫값이나 정수형의 슬라이싱, 팬시 리스트 값을 입력해야 함 / 열 위치에 -1을 입력하여 데이터프레임의 가장 마지막 열 데이터를 가져오는 데 자주 사용하기도 함
- iloc[:, -1] > 맨 마지막 칼럼만, iloc[:, :-1] > 맨 마지막 칼럼 제외 전부
- loc[행 위치 데이터프레임 인덱스 값, 열 위치 칼럼명]: label 기반 인덱싱 ex) 인덱스 값이 one인 행의 칼럼명이 name인 데이터 추출하려면 data_df.loc['one', 'Name'] / loc에서 슬라이싱할 때는 종료값 -1이 아니라 종료 값까지 포함됨, 명칭은 숫자형이 아닐 수 있기 때문 / iloc와 달리 불린인덱싱이 가능(iloc은 정수형이 아닌 불린값은 지원하지 않아서)
- [] 내에 불린인덱싱을 적용하면 반환되는 객체가 DataFrame이므로 원하는 칼럼명만 별도로 추출 가능

정렬

- 데이터프레임, 시리즈의 정렬: sort_values()
- by: 특정 컬럼 입력하면 해당 칼럼으로 정렬 수행
- ascending = True일 경우 오름차순, False일 경우 내림차순으로 정렬, 기본은 오름차순
- inplace = False로 설정 시 sort_values()를 호출한 데이터프레임은 그대로 유지하며 정렬된 데이터프레임을 결과로 반환함 / True로 설정하면 호출한 데이터프레임의 정렬 결과를 그대로 적용함

Aggregation 함수 적용
- min(), max(), sum(), count() 등의 함수를 aggregation 함수라 함 > null값을 반영하지 않아서 칼럼마다 count 결과값이 다를 수 있음
- 특정 칼럼에만 적용하고 싶을 경우 데이터프레임에 대상 칼럼들만 추출해서 함수 적용하면 됨

groupby() 적용
- 입력 파라미터 by에 칼럼을 입력하면 대상 칼럼으로 groupby 됨
- 데이터프레임에 groupby() 호출하면 DataFrameGroupBy라는 또 다른 형태의 DataFrame 반환
ex ) groupby(by = 'Pclass')를 호출하면 Pclass 칼럼 기준으로 GroupBy된 DataFrameGroupby 객체를 반환
- DataFrame에 groupby()를 호출해 반환된 결과에 aggregation 함수를 호출하면 groupby() 대상 칼럼을 '제외한' 모든 칼럼에 해당 aggregation 함수를 적용
- DataFrame의 groupby( )에 특정 칼럼만 aggregation 함수를 적용하려면 grouby( )로 반환된 DataFrameGroupBy 객체에 해당 칼럼을 필터링한 뒤 aggregation 함수를 적용해야 함. ex) titanic_df.groupby('Pclass')로 반환된 DataFrameGroupBy 객체에 [['Passengerld', 'Survived']]로 필터링해 Passengerld와 Survived 칼럼에만 count()를 수행.
- DataFrame groupby()의 경우, 적용하려는 여러 개의 aggregation 함수명을 DataFrameGroupBy 객체의 agg( ) 내에 인
자로 입력해서 사용.

결손 데이터 처리
- isna()로 결손 데이터 여부 확인: 모든 칼럼의 값이 NaN인지 아닌지를 True나 False로 알려 줌 > sum() 함수를 추가해서 결손 데이터의 개수를 구할 수 있음 ex) titanic_df.isna( ).sum( )
- fillna()로 결손 데이터 대체하기: 대체하고자 하는 값을 인자 안에 입력해 주면 칼럼 내 결손 데이터들이 모두 인자 값으로 바뀜, 반환 값을 반드시 다시 받거나 inplace=True 파라미터를 fillna('대체값', inplace = True)처럼 추가해야 실제 데이터 세트 값이 변경됨

apply lambda 식으로 데이터 가공
- 판다스는 apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하
는 기능을 제공함.
- lambda x : x ** 2에서 ‘:’로 입력 인자와 반환될 입력 인자의 계산식을 분리함. ‘:’의 왼쪽에 있는 x는 입력 인자, 오른쪽은 입력 인자의 계산식을 의미.
- lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우에는 squares = map(lambda x : x**2, a)와 같은 형태로 map() 함수를 결합해서 사용. 이때 a는 리스트 형식 등으로 주어질 수 있음
- titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x〈=15 else 'Adult') 처럼 if절과 lambda를 함께 사용할 수 있음, if식보다 반환 값을 먼저 기술해 줘야 함
반환값 다음에 if 조건 명시할 것, else if는 지원 x

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else: cat = 'Elderly'

  return cat

#lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정.
#get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
