2. 데이터를 불러옵니다.

```python
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
smpl_sub = pd.read_csv('/content/sample_submission.csv')
```

```python
train.head()
```

```python
test.head()
```

3. 변수를 변환합니다.

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.

#mapping 방법 이용
type_map = {'white': 0,
            'red': 1}

train['type'] = train['type'].map(type_map)
test['type'] = test['type'].map(type_map)
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.

train_x = train.drop(columns=['index', 'quality'])
train_y = train['quality']
test_x = test.drop(columns=['index'])
```

```python
train_x.shape, train_y.shape, test_x.shape

# train x에는 12개의 피처, 5497개의 데이터
# train y에는 데이터 5497개 존재
# test x에는 1000개의 데이터 12개의 피처
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
test_x.head()
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
from lightgbm import LGBMClassifier

lgbm_clf = lgbm.LGBMClassifier(random_state = 0)
lgbm_clf.fit(train_x, train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
preds = lgbm_clf.predict(test_x)

print(preds[:10])
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head()
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
smpl_sub['quality'] = preds
```

```python
smpl_sub.head()
```

```python
# csv 파일로 저장합니다.
smpl_sub.to_csv('submission.csv', index=False)
```

# 2. XGBoost

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
# 데이터 확인
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())

train = pd.get_dummies(train, columns=['type'])
test = pd.get_dummies(test, columns=['type'])
```

```python
train.head()

#white면 True, red면 false
```

```python
test.head()
```

```python
# 모델 정의

from xgboost import XGBClassifier
XGB = XGBClassifier(random_state = 0)
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수


# fit 메소드를 이용해 모델 학습

from sklearn.preprocessing import LabelEncoder

y = train['quality']
X = train.drop('quality', axis = 1)
# quality 값을 0부터 시작하도록 인코딩
le = LabelEncoder()
y = le.fit_transform(y)

XGB.fit(X, y)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측

preds = XGB.predict(test)
print(preds[:10])

#라벨 인코딩의 원본 예측값
original_preds = le.inverse_transform(preds)
print(original_preds[:10])
```

```python

# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
smpl_sub['quality'] = original_preds
```

```python
# 정답파일 내보내기
smpl_sub.to_csv('xgb_pred.csv', index=False)
```

# 3. 보팅

```python
# 데이터 다운로드 링크로 데이터를 코랩에 불러옵니다.

!wget 'https://bit.ly/3i4n1QB'

import zipfile
with zipfile.ZipFile('3i4n1QB', 'r') as existing_zip:
    existing_zip.extractall('data')
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train = pd.get_dummies(train, columns=['type'])
test = pd.get_dummies(test, columns=['type'])
```

```python
# 모델 정의 (LightGBM, XGBoost, Random Forest)

lgbm = LGBMClassifier(random_state = 0)
xgb = XGBClassifier(random_state = 0)
rf = RandomForestClassifier(random_state = 0)

# VotingClassifier 정의

voting_clf = VotingClassifier(estimators=[('lgbm', lgbm), ('xgb', xgb), ('rf', rf)], voting='hard')
```

```python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수

X = train.drop('quality', axis = 1)
y = train['quality']

# fit 메소드를 이용해 모델 학습
voting_clf.fit(X, y)
```

```python
# predict 메소드와 원핫 인코딩한 테스트 데이터를 이용해 품질 예측
pred = voting_clf.predict(test)
print(pred[:10])
```
