1. 데이터 전처리 & EDA
- 먼저 요일에 따른 18-20시 탑승객 수를 바플롯으로 시각화하여 요일에 따른 차이, 주말과 주중의 차이를 알아보고 요일별 해당 시간 탑승 인원의 평균과 합이라는 새로운 피처를 생성함 -> 새로운 변수를 시각화함으로써 인사이트를 도출해 낸 것이 인상 깊음
- 배차 간격에 따른 퇴근시간 탑승객 수 사이의 패턴을 명확히 보기 위해 로그변환 사용 -> 로그 변환을 사용하니 그래프를 알아보는 것이 더 용이하고 유의한 정보 얻을 수 있었음

- RandomForest
-> 부스팅 모델이 랜덤포레스트보다 일반적으로 성능이 높지만 과적합 위험이 크기 때문에 더 안전한 랜덤포레스트 선택: 모델의 특징을 파악하면 데이터에 맞는 모델을 설계할 때 도움이 됨.
- Ensemble
-> 모델 간 상관계수 행렬을 만들고, 상관관계 높은 모델은 역 평균을 수행하고 0과 가중 산술 평균을 수행함. 아직 내 수준에선 어려운 단계지만 후에 알아두면 도움이 될 것 같다.
