# 합성곱 신경망

앞선 내용: 순전파 과정에 따라 계산된 오차 정보가 신경망의 모든 노드(출력층 -> 은닉층 -> 입력층)로 전송됨 (역전파). 이는 복잡하고 동시에 많은 자원을 요구해서 시간 소요가 큼.

위의 문제를 해결하고자 등장한 것이 합성곱 신경망.

정의: 이미지 전체를 한번에 계산하는 것이 아니라, 국소적인 부분을 계산하여 시간과 자원을 절약하는 방식. 주로 이미지나 영상, 음성 인식 등에 유용함.

## 합성곱층의 필요성
이미지 분석은 3X3 같은 배열을 flattening 과정을 거쳐 각 픽셀에 가중치를 곱하여 은닉층으로 전달함. 그러나 이렇게 이미지를 flatten 하면 데이터의 공간적 구조를 무시하게 되는데, 이를 방지하고자 도입된 것이 합성곱층.

## 합성곱 신경망 구조

1. 입력층: 입력 이미지가 최초로 거치게 되는 계층. 입력 이미지는 너비, 높이, 채널의 값을 갖는 3차원 데이터로, 흑백 이미지면 채널은 1, 컬러면 3 값을 가짐.

----------아래 두 과정에서 입력 이미지 주요 특성 벡터 추출------------

2. 합성곱층: 입력 데이터 특성을 추출하는 역할 수행
- 커널/필터를 통해 이미지의 모든 영역을 훑으면서 특성을 추출. 이때 커널은 3X3. 5X5가 일반적이며, 스트라이드라는 지정된 간격에 따라 순차적으로 이동함. 순회하면서 모든 입력 값과의 합성곱 연산으로 새로운 특성 맵을 만들게 되며, 원본보다 작은 크기의 특성맵 출력.
- 특성맵: 추출된 특성 결과물
- 흑백 이미지는 필터의 채널이 1이고, 컬러 이미지는 채널이 3임.
- 컬러 이미지는 rgb 각각에 서로 다른 가중치로 합성곱을 적용한 후 결과를 더함. 그 외 스트라이드 및 연산 방법 동일 -> 채널이 3이라고 해서 필터 개수가 여러 개인 것이 아님. 필터는 실제로 하나임을 주의.
- 필터가 두 개 이상인 합성곱은, 필터 각각이 특성 추출 결과의 채널이 됨.

요약
- 입력 데이터는 가로(W), 세로(H), 채널(또는 깊이 D) -> W*H*D
- 필터 개수 K, 필터 크기 F, 스트라이드 S, 패딩 P
- 출력 데이터: W2 = (W-F+2P) / S+1, H2 = (H-F+2P)/S+1, D = K

3. 풀링층: 특성 맵의 차원을 다운 샘플링, 즉 축소하여 연산량을 감소시키고 주요한 특성 벡터를 추출하여 학습을 효과적으로 할 수 있게 함.

- 최대 풀링: 대상 영역에서 최댓값 추출, 일반적으로 많이 사용
- 평균 풀링: 대상 영역에서 평균을 반환, 커널 값을 평균화시키는데 이 과정에서 중요한 가중치를 갖는 특성이 희미해질 수 있음

요약
- 입력 데이터는 가로(W), 세로(H), 채널(또는 깊이 D) -> W*H*D
- 필터 크기 F, 스트라이드 S
- 출력 데이터: W2 = (W-F) / S+1, H2 = (H-F)/S+1, D = D

----------------------------------
4. 완전연결층: 특성 벡터들이 완전 연결층을 거치며서 1차원 벡터로 변환됨

- 이 과정에서 이미지는 3차원 벡터에서 1차원 벡터로 펼처지게 됨 flatten

5. 출력층: 소프트맥스 함수를 사용하여 최종 결과 출력

- 이미지가 각 레이블에 속할 확률 값이 출력되며 이때 가장 높은 확률 값을 갖는 레이블이 최종 값으로 선정됨

## 1D. 2D, 3D 합성곱

1. 1D 합성곱: 필터가 시간을 축으로 좌우로만 이동 가능한 합성곱. 입력 W와 필터 K에 대한 출력은 W가 되는데, 출력값의 요소 개수는 필터 요소 수와 동일. 그래프 곡선 완화할 때 많이 사용됨.

2. 2D 합성곱: 필턱 두 가지 방향으로, 위아래와 좌우로 움직이는 형태. 입력 (W,H)와 필터(k,k)에 대한 출력은 (W,H)가 되며 출력 형태는 2D 행렬이 됨.

3. 3D 합성곱: 필터가 대각선으로까지 움직이는 것이 가능함. 입력 (W,H,L)에 대해 필터(k,k,d)를 적용하면 출력으로 (W,H,L)을 갖는 형태가 3D 합성곱. 출력은 3D 형태며, d<L 유시하는 것이 중요

4. 3D 입력을 갖는 2D 합성곱: 입력은 3D 형태임에도 출력 형태가 2D 행렬을 취하는 것. 필터에 대한 길이 L이 입력 채널의 길이 L과 같아야 하기 때문에 이와 같은 합성곱 형태가 만들어진 것. 입력 (W,H,L)에 필터 (k,k,L)를 적용하면 (W, H) 출력. 이때 움직이는 방향은 2D와 동일

5. 1x1 합성곱: 3D 형태로 입력되며, 입력 (W,H,L) 필터 1,1,L을 적용하면 출력은 W,H가 됨. 채널수를 조정해서 연산량 감소 효과 존재
